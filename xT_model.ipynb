```python
from statsbombpy import sb
def get_xT_data (competition_id = 43, season_id = 3):
    matches = sb.matches(competition_id = competition_id , season_id = season_id)
    all_xT_events = []
    for match_id in matches['match_id']:
        events = sb.events(match_id= match_id)
        valid_events = ['Pass','Carry','Shot']
        xT_events = events[events['type'].isin(valid_events)].copy()
        all_xT_events.append(xT_events)
    return pd.concat(all_xT_events,ignore_index = True)
xt_df = get_xT_data()
```

```python

def get_first_value(val_str):
    """Extracts the first value from a string like '[70.0,40.0]'"""
    if isinstance(val_str, str):
        # Remove brackets and split by comma
        parts = val_str.strip('[]').split(',')
        # Convert the first part to float, then int
        return int(float(parts[0]))
    return None # Returns None if the row is missing data

def get_second_value(val_str):
    """Extracts the second value from a string like '[70.0,40.0]'"""
    if isinstance(val_str, str):
        # Remove brackets and split by comma
        parts = val_str.strip('[]').split(',')
        # Convert the second part to float, then int
        return int(float(parts[1]))
    return None # Returns None if the row is missing data
```

```python
df['x'] = df['location'].apply(get_first_value)
df['y'] = df['location'].apply(get_second_value)
```

```python
df[['x','y']]
```

```python
pitch_length = 120.0
pitch_width = 80.0
x_bins = 16
y_bins = 12

x_cell_size = pitch_length/x_bins
y_cell_size = pitch_width/y_bins

df['x_cell'] = np.clip((df['x']//x_cell_size).astype(int),0,x_bins-1)
df['y_cell'] = np.clip((df['y']//y_cell_size).astype(int),0,y_bins-1)
```

```python
df['pass_end_x'] = df['pass_end_location'].apply(get_first_value)
df['pass_end_y']= df['pass_end_location'].apply(get_second_value)
```

```python
df['carry_end_x']  = df['carry_end_location'].apply(get_first_value)
df['carry_end_y'] = df['carry_end_location'].apply(get_second_value)
```

```python
df['end_x'] = df['pass_end_x'].fillna(df.get('carry_end_x',np.nan))
df['end_y'] = df['pass_end_y'].fillna(df.get('carry_end_y',np.nan))
```

```python
mask = df['end_x'].notna()
df.loc[mask,'end_cell_x'] = np.clip((df.loc[mask,'end_x']//x_cell_size).astype(int),0,x_bins-1)
df.loc[mask,'end_cell_y'] = np.clip((df.loc[mask,'end_y']//y_cell_size).astype(int),0,y_bins-1)
```

```python
successful_moves = df[
    (df['type']=='Carry') | ((df['type']=='Pass') & (df['pass_outcome'].isna()))
].copy()
```

```python
transition_counts = successful_moves.groupby(
    ['x_cell','y_cell','end_cell_x','end_cell_y']
).size().reset_index(name='count')
```

```python
total_moves_from_cell = transition_counts.groupby(['x_cell','y_cell'])['count'].transform('sum')
```

```python
transition_counts['transition_probability'] = transition_counts['count'] / total_moves_from_cell
print(transition_counts)
```

```python
transition_matrix = np.zeros((16,12,16,12))
for index , row in transition_counts.iterrows():
    sx = int(row['x_cell'])
    sy = int(row['y_cell'])
    ex = int(row['end_cell_x'])
    ey = int(row['end_cell_y'])

    prob = row['transition_probability']

    transition_matrix[sx,sy,ex,ey] = prob

sum_probs = np.sum(transition_matrix[13,5,:,:])
print(f"Sum of transition probabilites from cell (13,5): {sum_probs:.2f}")
```

```python
p_shot_grid = np.zeros((16,12))
p_move_grid = np.zeros((16,12))
xG_grid = np.zeros((16,12))

for x in range(16):
    for y in range(12):
        cell_events = df[(df['x_cell']==x) & (df['y_cell']==y)]
        total_events = len(cell_events)
        if total_events ==0 :
            continue
        shots = cell_events[cell_events['type']=='Shot']
        successful_moves = cell_events[
            (cell_events['type'] == 'Carry') | 
            ((cell_events['type'] == 'Pass') & (cell_events['pass_outcome'].isna()))
        ]
        p_shot_grid[x,y] = len(shots)/total_events
        p_move_grid[x,y] = len(successful_moves)/total_events

        if len(shots) > 0:
            goals = shots['shot_outcome']=='Goal'
            xG_grid[x,y] = goals.mean()
print("Grids build succesfully!")

print(f"Zone 14 - P(Shoot) : {p_shot_grid[13,5]:.3f}")
print(f"Zone 14 - P(move) : {p_move_grid[13,5]:.3f}")
print(f"Zone 14 - Avg xG : {xG_grid[13,5]:.3f}")
```

```python
xT_grid = np.zeros((16,12))
num_iterations = 50
tolerance = 0.0001

print("Starting value iteration")
for i in range(num_iterations):
    new_xT_grid = np.zeros((16,12))
    max_change = 0
    for x in range(16):
        for y in range(12):
            immediate_threat = p_shot_grid[x,y]*xG_grid[x,y]
            future_threat = np.sum(transition_matrix[x,y,:,:]*xT_grid)
            new_value = immediate_threat + (p_move_grid[x,y]*future_threat)
            new_xT_grid[x,y] = new_value
            change = abs(new_value - xT_grid[x,y])
            if change > max_change:
                max_change = change
    xT_grid = new_xT_grid.copy()
    if max_change < tolerance:
        print(f"Converged nicely after {i+1} iterations")
        break
print("Expect threat model complete")
```

```python
import pandas as pd
import numpy as np

def calculate_player_xt(df, player_name, xt_grid):
    """
    Calculates the total Expected Threat (xT) added by a specific player.
    """
    player_events = df[df['player'] == player_name].copy()
    
    if len(player_events) == 0:
        return f"No events found for {player_name}."

    successful_actions = player_events[
        (player_events['end_cell_x'].notna()) &  # Must have an end location
        (
            (player_events['type'] == 'Carry') | 
            ((player_events['type'] == 'Pass') & (player_events['pass_outcome'].isna()))
        )
    ].copy()
    
    if len(successful_actions) == 0:
        return f"{player_name} had no successful passes or carries."

    start_x = successful_actions['x_cell'].astype(int).values
    start_y = successful_actions['y_cell'].astype(int).values
    end_x = successful_actions['end_cell_x'].astype(int).values
    end_y = successful_actions['end_cell_y'].astype(int).values

    start_xt_values = xt_grid[start_x, start_y]
    end_xt_values = xt_grid[end_x, end_y]
    successful_actions['xT_added'] = end_xt_values - start_xt_values
    
    total_xt = successful_actions['xT_added'].sum()
    total_actions = len(successful_actions)

    print(f"--- Scouting Report: {player_name} ---")
    print(f"Successful Progressions: {total_actions}")
    print(f"Total xT Added: {total_xt:.3f}")
    print(f"xT per Action: {(total_xt / total_actions):.4f}")
    
    return successful_actions

messi_xt_df = calculate_player_xt(df, "Lionel Andrés Messi Cuccittini", xT_grid)
```
