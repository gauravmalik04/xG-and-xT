```python

from statsbombpy import sb 
import pandas as pd

def get_shot_data(competition_id = 43 , season_id = 3):
    matches = sb.matches(competition_id = competition_id, season_id = season_id)

    all_shots = []
    for match_id in matches['match_id']:
        events = sb.events(match_id = match_id)
        shots = events[events['type']=='Shot'].copy()
        all_shots.append(shots)
    return pd.concat(all_shots)

df = get_shot_data()
df[['x','y']] = pd.DataFrame(df['location'].tolist(),index = df.index)
print(f"Loaded {len(df)} shots for analysis.")
```

```python
import numpy as np
df['distance_from_goal'] = np.sqrt((120-df['x'])**2 + (40-df['y'])**2)
print(df[['x','y','distance_from_goal']].head())
```

```python
import numpy as np
goal_width = 8
left_post_y = 36
right_post_y = 44

d1 = np.sqrt((120-df['x'])**2 + (left_post_y-df['y'])**2)
d2 = np.sqrt((120-df['x'])**2 + (right_post_y-df['y'])**2)

cos_theta = (d1**2  + d2**2  - goal_width**2) / (2*d1*d2)

# clipping so that the anlge remaining the range of arccos i.e. cos inverse as it's range is between -1 to 1, and due to decimal arithmatic being error prone a value such as 1.0000000000002 may crash the code

cos_theta = np.clip(cos_theta,-1.0,1)


df['shot_angle'] = np.degrees(np.arccos(cos_theta))
print(df[['distance_from_goal','shot_angle']].head())
```

```python
# statsbomb stores it body part data as a dictionary such as {'id':40,'name':'Right Foot'}

df_encoded = pd.get_dummies(df,columns = ['shot_body_part'],prefix="shot_type")
# print(df_encoded.columns)


print(df_encoded[['distance_from_goal','shot_angle','shot_type_Head','shot_type_Right Foot','shot_type_Left Foot']].head())
```

```python
df_encoded['is_goal'] = np.where(df['shot_outcome']=='Goal',1,0)
print(df_encoded['is_goal'].value_counts(normalize =True)*100)
```

```python
from sklearn.metrics import brier_score_loss , log_loss, roc_auc_score

def evaluate_xg_model(y_true,y_pred_proba):
    """
    Evaluates an xG model's probabilistic predictions.
    y_true = actual binary outcome(0 or 1).
    y_pred_proba : predicted probabilities (0.0 to 1.0)
    """
    brier = brier_score_loss(y_true,y_pred_proba)
    # Log loss requires careful handling of absolute of 0s and 1s, which sklearn handles safely
    ll = log_loss(y_true , y_pred_proba)

    ## ROC-AUC tells us how well the model ranks shots
    # E.g. does a 0.8 xG shot genuinely look better than a 0.2 xG shot?
    auc = roc_auc_score(y_true , y_pred_proba)
    print(f"--- xG Model Evaluation ---")
    print(f"Brier Score: {brier:.4f} (Closer to 0 is better)")
    print(f"Log Loss:    {ll:.4f} (Closer to 0 is better)")
    print(f"ROC-AUC:     {auc:.4f} (Closer to 1.0 is better)")
    return brier,ll,auc
# example usage : 
y_true = np.array([0,0,1,0,1])
y_pred = np.array([0.05,0.12,0.88,0.30,0.15])
evaluate_xg_model(y_true,y_pred)
```

```python
# Building a baseline logistic regression model
from sklearn.linear_model import LogisticRegression
features = ['distance_from_goal','shot_angle','shot_type_Right Foot','shot_type_Head','shot_type_Left Foot']
X = df_encoded[features]
y = df_encoded['is_goal']
```

```python
lr_model = LogisticRegression(solver='liblinear',max_iter=1000)
lr_model.fit(X,y)
```

```python
df_encoded['baseline_xG'] = lr_model.predict_proba(X)[:,1]
df_encoded.to_csv("shots_data_refactored")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train,X_test,y_train,y_test = train_test_split(
    X,y,test_size=0.2,stratify=y,random_state=42
    )
```

```python
print(f"Training_set : {len(X_train)} shots. Goal Ratio : {y_train.mean():.3f}")
print(f"Testing set: {len(X_test)} shots. Goal Ratio: {y_test.mean():.3f}")
```

```python
startified_model = LogisticRegression(solver='liblinear' , max_iter = 1000)
startified_model.fit(X_train,y_train)
```

```python
y_pred_proba_stratified = startified_model.predict_proba(X_test)[:,-1]
```

```python
evaluate_xg_model(y_test,y_pred_proba_stratified)
```

- A linear model sees a negative weight for distance and applies it uniformly. It says: "If 10 meters is good, 5 meters is better, and 60 meters is basically impossible." It cannot easily account for the fact that a world-class striker might have a specific "sweet spot" at 25 yards where they are actually quite dangerous.

### Looking at the weights learned by the model

```python
weights = pd.DataFrame({
    'feature' : features,
    'Weight' : startified_model.coef_[0]
})
print(weights.sort_values(by='Weight',ascending=False))
```
